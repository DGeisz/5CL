```python
data = pd.read_csv('vernierData.csv')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot(data.Lpix, data.spix)
plt.show()
```

```python
slope =  (600) / (980 - 328)
data2 = pd.DataFrame()
data2['Better'] = slope*(data.Lpix - 328) + 200
data2
```

```python
slope2 = (0.4) /(886 - 499) 
data2['Betterer'] = slope2*(886 - data.spix)
```

```python
data2
```

```python
plt.plot(data2.Better, data2.Betterer)
```

```python
from scipy.interpolate import interp1d
f = interp1d(data2.Better, data2.Betterer)
```

```python
xnew = np.linspace(450, 700, 100)
plt.plot(xnew, f(xnew))
plt.show()
```
