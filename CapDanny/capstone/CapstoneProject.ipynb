# Theory

For our capstone project, we wish to test Planckâ€™s Law for the Blackbody Radiation spectrum.  Planck's Law takes the form:

$$B_\lambda (\lambda, T) = \frac{2hc^2}{\lambda ^5}\frac{1}{e _{}^{\frac{hc}{\lambda k_B T} - 1}}$$

Here, $B_{\lambda}$ is the spectral radiance of a Blackbody, $\displaystyle h$, $c$, and $k_B$ are Planck's unreduced constant, the speed of light, and Boltzmann's Constant respectively, and $T$ and $\lambda$ are the temperature of Blackbody and the wavelength of light respectively.


We also wish to verify Wien's Law.  Wien's Law states that the wavelength $\lambda_{max}$ at which the spectral radiance of a Blackbody is greatest is related to the Blackbody Temperature $T$ according to the relation:

$$\lambda_{max} = \frac{b}{T}$$
Where $b$ is Wien's Constant and theoretically takes the form:

$$b = \frac{hc}{xk_B}$$

where $x$ is the solution to the equation:

$$\frac{xe^x}{e^x - 1} - 5 = 0$$

which can be numerically solved to yield:

$$x \approx 4.965114231744276$$

In this lab, we will be using a tungsten bulb as our approximate black body.  In order to determine the temperature of the tungsten filament, we are going to use the temperature dependence of resistivity formula for metals, which takes the form:

$$R_T = R_0 \left[1 + \alpha\left(T - T_0\right)\right]$$

$$\implies T = \left(\dfrac{R}{R_{0}} - 1\right) \dfrac{1}{\alpha} + T_{0}$$     
        

Where $R_0$ is the resistance of the metal at temperature $T_0$, $R_T$ is the temperature of the metal at temperature $T$, and $\alpha$ is a constant that depends of the metal.  The final equation gives the error propagation formula. Using Ohm's law,

$$R = \frac{V}{I}$$

we will determine the resistance of the filament at room temperature.  To determine the temperature of the filament when we power the light bulb, we can use Ohm's law to determine the resistance of the filament, and then use the fact that $\alpha = 0.0045 \frac{Ohm}{^0C}$ for tungsten to determine the temperature of the filament.  Using $V_0 = I_0 R_0$ and $V = IR$, the corresponding formula for the error propagation of the above formula is given by:

$$\delta T = \sqrt{\left(\dfrac{I_{0}\delta V}{I\alpha V_{0}}\right)^2 + \left(\dfrac{VI_{0}\delta I}{V_{0}\alpha I^{2}}\right)^{2} + \left(\dfrac{V\delta I_0}{IV_{0}\alpha}\right)^{2} + \left(\dfrac{VI_{0}\delta V_0}{I\alpha V_{0}^{2}}\right)^{2}}$$

## Procedure

1. Shine the first of the three lasers through the opening to the monochromator.
2. Adjust the wavelength dial on the monochromator until the laser light is emitted from the device.
3. Record the monochromator's wavelength reading and the actual wavelength of the laser.
4. Repeat steps 1-3 with the other two lasers.  
5. Use the data collected to determine the offset between the monochromator's wavelength reading and the wavelength of light actually being emitted from the device.

### Experimental Procedure
1. Set up materials according the diagram below. 
![Apparatus](IMG-0012.JPG)
2. Turn on the Tungsten bulb to a low setting.
3. Take the lid off the monochromator, and align the bulb so that the light emitted into the monochromator is incident on both mirrors and the grating inside the monochromator. Do not adjust any of the optical devices within the monochromator.
4. Align the photodetector so that the light emitted from the monochromator is incident on the aperture of the detector.
5. The smallest reading from a vernier photodetector is 2.0 Lux, so adjust the wavelength dial on the monochromator until the photodetector reads a value just above 2.0 Lux.

In our case, we started at the smallest wavelength for which the photodetector read a value greater than 2.0 Lux, and we increased the wavelength by about 100 Angstrom between subsequent data points.
7. Repeat steps 5 and 6 four times, but each time change the power supplied to the Tungsten bulb to change its temperature.  Be sure to record the power setting each time you repeat the experiment.
8. Turn off the power supply, and connect the multimeter in parallel to the circuit, and clamp the current clamp onto one of the light bulb wires.
9. Turn on the power supply to the lowest setting at which current flows through the bulb.
10. Use the multimeter and the current clamp to find the voltage across the light bulb and the current running through the circuit.  Use these values to give an approximate value of the resistance of the bulb at room temperature.
11. Turn the power supply to each of the power settings used in steps 1-7, and at each setting, record the voltage across the light bulb and the current in the circuit. Used with equations in the theory section, these values can give you an approximate value for the temperature of the filament.

We therefore see that this data could potentially be modeled by a linear hypothesis.  If we fit the data to the hypothesis: 

$$\lambda_{dmax} = \hat{m}\left(\frac{1}{T}\right) + \hat{b}$$

We would hope to find that $\hat{b} = 0$, which would show that $\lambda_{dmax}$ is in fact a constant multiple of $\lambda_{max}$, and would therefore somewhat validate Wien's Law.  Because uncertainty for our values of temperature are humongous, we will use a weighted least-squares fit for this data.  The script that calculates $\hat{m}$ and $\hat{b}$ is quite lengthy so I have included it in the references section.

As can be seen in the references section, we have that

$$\hat{m} = 1245757.3116098575 \pm 4742295.84971805\;^oC \cdot \unicode[serif]{xC5}$$
$$\hat{b} = 5597.622005875043 \pm 2058.7735784622555\;\unicode[serif]{xC5}$$

A simple agreement test between our value of $\hat{b}$ and the expected value of 0 yields 1.359, so our experimentally determined value of $\hat{b}$ does agree with what we would expect from Wien's Law.  This indicates that the maxima we see in our data is not related to Wien's law, but rather probably has to do with an errors in our data for the photodetector's spectral sensitivity or the monochromator's spectral efficiency.  

At this point we have seen that our data does not resemble the theoretical prediction of Planck's law, and we have also seen that our data also does not support Wien's Law.  Perhaps the only piece of theoretical physics that our data even slightly supports is the Stefan-Boltzmann Law, which states that the radiant emittance $j$ of a blackbody is related to temperature according to the relation:

$$j = \sigma T^4$$

Where $\sigma$ is Stefan-Boltzmann constant.  While the nature of our experimental apparatus is such that we don't have the right data to properly compute $j$, we generally expect that as we increase temperature, spectral radiant intensity should increase over all wavelengths. The following script creates a numpy interpolation object (basically a function) that represents the radiant sensitivity of the Photodetector.

```python
p_photo = pd.DataFrame()
photo_data = pd.read_csv('vernierData.csv')
Lslope = 600 / (980 - 328)
p_photo['lambda'] = Lslope * (photo_data.Lpix - 328) + 200
Sslope = 0.4 / (886 - 499)
p_photo['rad_sens'] = Sslope * (886 - photo_data.spix)
photo_sensitivity = interp1d(p_photo['lambda'], p_photo.rad_sens)
```

The following scripts creates a numpy interpolation object (basically a function) that represents the spectral efficiency of the monochromator.

```python
m_sens = pd.DataFrame()
m_data = pd.read_csv('monochromator.csv')
m_sens['lambda'] = (500 / (982 - 272)) * (m_data.wlPix - 272) + 300
m_sens['eff'] = (1 / (526 - 244)) * (526 - m_data.effPix)
monochromator_sensitivity = interp1d(m_sens['lambda'], m_sens.eff)
```
